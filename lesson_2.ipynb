# 1. Linear Classification

## Line Equation

The line equation is defined as $y = mx + b$, where $m$ is the slope and $b$ is the intercept. The slope is the rate of change of the line, or in other words, how steep the line is. The intercept is the point where the line crosses the y-axis. While studying Neural Networks, we will represent the line equation as $w_1x_1 +w_2x_2 + b$, where $w_1, w_2, b$ are called parameters that define the linear model (the line).

Plotting an example line equation: $-1x_1 + 4x_2 + 0.4$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_line(a, b, c):
    x = np.linspace(-2, 4, 50)
    y = (-a*x - c)/b
    
    plt.axvline(0, color='black')
    plt.axhline(0, color='black')
    plt.plot(x, y)
    plt.grid(True)
```

```python
a = -1
b = 4
c = 0.4

plot_line(a, b, c)
```

## Solving the equation for different points

```python
p1 = (2, 0.4)
ret = a*p1[0] + b*p1[1] + c

print("%.2f" % ret)
```

```python
plot_line(a, b, c)
plt.plot(p1[0], p1[1], 'ro')
```

```python
p2 = (1, 1)
ret = a*p2[0] + b*p2[1] + c

print("%.2f" % ret)
```

```python
plot_line(a, b, c)
plt.plot(p2[0], p2[1], 'go')
```

```python
p_3 = (2, -0.2)
ret = a*p_3[0] + b*p_3[1] + c

print("%.2f" % ret)
```

```python
plot_line(a, b, c)
plt.plot(p_3[0], p_3[1], 'bo')
```

Observing the points plotted above, we can see that the points above the line have a positive value, while the points below the line have a negative value. This is the basis of linear classification. Given a line equation $f(x)$, we can classify a point $x$ as positive if $f(x) > 0$ and negative if $f(x) < 0$.

# Linear Model

A linear model is a mapping function $X \rightarrow Y$ that can be represented as $f(x) = w_1x_1 + w_2x_2 + ... + w_nx_n + b$, where $w_1, w_2, ..., w_n$ are the weights and $b$ is the bias. The weights and bias are the parameters of the model. The weights are the coefficients of the linear model, while the bias is the intercept. The weights and bias are learned during the training process.
