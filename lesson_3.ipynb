```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np
```

# 1. Plotting the data points

We're using a dataset that has 2 features and 2 classes. We'll plot the data points and see how they are distributed.

```python
X, Y = make_classification(n_features=2, n_redundant=0, n_informative=1, n_clusters_per_class=1, random_state=46)
```

```python
plt.scatter(X[:,0], X[:,1], marker='o', c=Y, edgecolors='k')
```

# 2. Creating a model

Now that we have the data scattered, we shall plot a line that better separates the two classes. We'll use the equation of a line to do so. The equation of a line is given by:

$$
y = w_1x_1 + w_2x_2 + b
$$

where $w_1$ and $w_2$ are the weights and $b$ is the bias. We'll use the following values for the weights and bias:

$$
w_1 = 1, w_2 = 5, b = -0.4
$$

```python
def plot_model(w1, w2, b):
    plt.scatter(X[:,0], X[:,1], marker='o', c=Y, edgecolors='k')
    
    xmin, xmax = plt.gca().get_xlim()
    ymin, ymax = plt.gca().get_ylim()
    
    x = np.linspace(-2, 4, 50)
    y = (-w1 * x - b) / w2

    plt.axvline(0, color='black')
    plt.axhline(0, color='black')
    plt.plot(x, y)
    plt.grid(True)
    
    plt.xlim(xmin, xmax)
    plt.ylim(ymin, ymax)
```

```python
plot_model(5, 1, -0.4)
```

# 3. Classifying the data points

To classify the data points, we'll use the following function:

```python
def classify(point, w1, w2, b):
    ret = w1 * point[0] + w2 * point[1] + b
    
    if ret >= 0:
        return 1, 'yellow'
    else:
        return 0, 'white'
```

```python
p = (2, -1)

_class, color = classify(p, 5, 1, -0.4)

print('Class: ', _class)
print('Color: ', color)
```

```python
plot_model(5, 1, -0.4)
plt.plot(p[0], p[1], marker='^', c=color, markersize=30)
```

# 4. Analyzing the model's accuracy

Now that we have a model, we'll see how accurate it is. We'll use the following function to calculate the accuracy of the model:

$$
accuracy = \frac{correct\ predictions}{total\ predictions}
$$

where $correct\ predictions$ is the number of data points that were correctly classified and $total\ predictions$ is the total number of data points.

```python
right_predictions = 0
for i in range(len(X)):
    _class, _ = classify(X[i], 5, 1, -0.4)
    
    if _class == Y[i]:
        right_predictions += 1

accuracy = right_predictions / len(X)

print('Accuracy: ', accuracy)
```
